# Complexity test

```python
import time
import matplotlib.pyplot as plt

node_count = [50, 250, 500, 1000, 2000, 3000, 4000, 5000]
alg_time = []

for i, n in enumerate(node_count):
    tree = generate_random_tree(n)
    start = time.time()
    result = solve_mis(tree)
    end = time.time()
    alg_time.append(end - start)

plt.plot(node_count, alg_time, 'bo-')
plt.xlim(left=min(node_count))
plt.xlabel("Nodes")
plt.ylabel("Time (s)")
plt.title("MIS Solution Time vs Tree Size")
plt.grid(True)
plt.show()
```
