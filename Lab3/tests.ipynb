# Base praph examples

## Path of 5 vertices

```python
graph = GRAPHS['path5']
nodes = sorted(graph.keys())
edges = [(u, v) for u in nodes for v in graph[u] if u < v]
print(f"Graph : {len(nodes)} vertices, {len(edges)} edges")
print(f"Edges : {edges}")

mis, _ = run_luby(graph)
```

```python
print(f"\nMaximal Independent Set")
print(f"  Size  : {len(mis)}")
print(f"  Nodes : {mis}")

visualize(graph, mis, title='Path of 5 vertices')
```
