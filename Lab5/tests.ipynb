```python
import math

from src.utils import DIRECTED_GRAPHS, GRAPHS, generate_random_directed_graph, generate_random_undirected_graph
from src.shortest_path import shortest_path, visualize_directed_graph
from src.graph_coloring import color_graph_parallel, visualize_undirected_graph
```

# Task 1. Shortest path

```python
source = '0'
```

## Small graph

```python
graph = DIRECTED_GRAPHS['small']

vertices = list(graph.keys())
edges = [(u, v, c) for u, neighbours in graph.items() for v, c in neighbours.items()]
print(f"Nodes  : {vertices}")
print(f"Edges  : {edges}")
print(f"Source : {source}")
```

```python
D = shortest_path(graph, source)

print("Shortest distances from", source)
for v in vertices:
    print(f"  D[{v}] = {'inf' if D[v] == math.inf else str(D[v])}")

visualize_directed_graph(graph, source, D, title=f"Small graph")
```

# Task 2 - Graph Coloring Minimization

## Triangle

```python
graph = GRAPHS['triangle']

vertices = list(graph.keys())
edges = [(u, v) for u, nbrs in graph.items() for v in nbrs if u < v]
print(f"Nodes  : {vertices}")
print(f"Edges  : {edges}")
print(f"Source : {source}")
```

```python
colors, rounds = color_graph_parallel(graph)

num_colors = max(colors.values())
print(f"Converged in {rounds} round(s), {num_colors} color(s) used")
print()
print("Final coloring:")
for v in vertices:
    print(f"  vertex {v} → color {colors[v]}")

visualize_undirected_graph(graph, colors, title=f"Graph coloring — triangle")
```
